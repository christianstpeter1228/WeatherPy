```python
# Dependencies and Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from datetime import datetime
from random import uniform
from citipy import citipy

# Import API key
api_key = "eb39b31f672ce3262434283aa0b0463a"
# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
columns = ["City", "Country"]
cities_df = pd.DataFrame(columns=columns)
```

## Generate Cities List

```python
for x in range(1000):
    lat, lng = np.random.uniform(low=-90.000, high=90.000), np.random.uniform(low=-180.000, high=180.000)
    city = citipy.nearest_city(lat, lng)
    cities_df = cities_df.append({"City": city.city_name,"Country": city.country_code,}, ignore_index=True)

cities_df = cities_df.drop_duplicates(subset='City').reset_index()
cities_df.count()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
columns = ["City", "Country","Temperature","Latitude","Longitude","Humidity","Cloudiness", "Wind Speed"]
clean_df = pd.DataFrame(columns=columns)
url = "http://api.openweathermap.org/data/2.5/weather?"

row_count = 1

print("Beginning Data Retrieval")
for index, row in cities_df.iterrows():
    city_row = row["City"]
    city_url = url + "appid=" + api_key + "&units=IMPERIAL" + "&q=" + city_row.replace(" ","+")
    city_data = requests.get(city_url).json()
    if city_data["cod"] == "404":
        print("City not found, skipping...")

    else:
        clean_df.set_value(index, "City", city_data["name"])
        clean_df.set_value(index, "Country", city_data["sys"]["country"])
        clean_df.set_value(index, "Temperature", city_data["main"]["temp"])
        clean_df.set_value(index, "Latitude", city_data["coord"]["lat"])
        clean_df.set_value(index, "Longitude", city_data["coord"]["lon"])
        clean_df.set_value(index, "Humidity", city_data["main"]["humidity"])
        clean_df.set_value(index, "Cloudiness", city_data["clouds"]["all"])
        clean_df.set_value(index, "Wind Speed", city_data["wind"]["speed"])
        
        print("------------------------")
        print("Processing: City # " , row_count, city_data["name"], city_data["sys"]["country"])
        print(city_url)
        row_count += 1
```
